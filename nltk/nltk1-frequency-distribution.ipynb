```python
import nltk
from nltk.probability import FreqDist
```

```python
lines=[line for line in open("data/quora-q1.txt")]
#print(lines)
text="".join(lines)
#print(text)
#sentences = nltk.tokenize.sent_tokenize(text)
#sentences_tokens = [nltk.tokenize.word_tokenize(s) for s in sentences]
tokens=nltk.tokenize.word_tokenize(text)
#print(tokens)
fd = FreqDist(tokens)
```

```python
print("Most frequent 20 words")
print(fd.most_common(20))
```

```python
print("Tabulate")
print(fd.tabulate)
```

```python
print("Occurrences of shoes")
print(fd['shoes'])
```

```python
print("Plot the most frequent 20 words")
fd.plot(20, cumulative=False)
```
