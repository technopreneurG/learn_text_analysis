A simple example to show how to load a text file, tokenise it, and use the Frequency Distribution class to get some interesting facts.

```python
import nltk
from nltk.probability import FreqDist
```

```python
lines=[line for line in open("data/quora-q1.txt")]
text="".join(lines)
tokens=nltk.tokenize.word_tokenize(text.lower())
#frequency distribution
fd = FreqDist(tokens)

#frequency distribution of counts
fdc = FreqDist(len(w) for w in tokens)

#calculate lexical richness of the text
def lexical_diversity(tokens):
    return len(set(tokens))/len(tokens)
```

```python
print("Total word count: {}".format(len(tokens)))
print("Unique word count: {}".format(len(set(tokens))))
print("Lexical diversity: {}".format(lexical_diversity(tokens)))
```

```python
print("Most frequent 20 words")
print("word, occurances, frequency")
for fw in fd.most_common(20) :
    print("'{}', {}, {:.2f}".format(fw[0], fw[1], fd.freq(fw[0])*100))
```

```python
print("Tabulate")
print(fd.tabulate)
```

```python
print("Occurrences of shoes")
print(fd['shoes'])
```

```python
print("Plot the 20 most frequent words")
fd.plot(20, cumulative=False)
```

```python
print("Most frequent 20 words lenghts")
print("word lengths, word count, frequency")
for fw in fdc.most_common(20) :
    print("'{}', {}, {:.2f}".format(fw[0], fw[1], fdc.freq(fw[0])*100))
```

```python
print("Plot the 20 most frequent word lengths")
fdc.plot(20, cumulative=False)
```

```python
#Frequency Distribution after removing stopwords
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
mystopwords = [",",".",")","(",":","","?","*","--","...","!","I","``","'s","'re"]
stopwords+=mystopwords
newtokens = [word for word in tokens if word not in stopwords]
#frequency distribution
fds = FreqDist(newtokens)
#frequency distribution of counts
fdsc = FreqDist(len(w) for w in newtokens)
```

```python
print("Total word count: {}".format(len(newtokens)))
print("Unique word count: {}".format(len(set(newtokens))))
print("Lexical diversity: {}".format(lexical_diversity(newtokens)))
```

```python
print("Most frequent 20 words")
for fw in fds.most_common(20) :
    print("'{}', {}, {:.2f}".format(fw[0], fw[1], fds.freq(fw[0])*100))
```

```python
print("Tabulate")
print(fds.tabulate)
```

```python
print("Occurrences of shoes")
print(fds['shoes'])
```

```python
print("Plot the 20 most frequent words")
fds.plot(20, cumulative=False)
```

```python
print("Most frequent 20 words lenghts")
print("word lengths, word count, frequency")
for fw in fdsc.most_common(20) :
    print("'{}', {}, {:.2f}".format(fw[0], fw[1], fdsc.freq(fw[0])*100))
```

```python
print("Plot the 20 most frequent word lengths")
fdsc.plot(20, cumulative=False)
```
